# This is Notebook shows basic  visualization on Koverse data using Pandas

## Start by connecting to Koverse throught the client library

```python
from koverse import client
client.connect()
```

# Then using Pandas, matplotlib, and Seaborn we can start visualizing data

## Query the Employees table for the Business Unit 'PB' select top 10 Organization Codes and put into a chart

```python
client.query_as_pandas_dataframe({'businessUnit':'PB'}, 'employees')['orgCode'].value_counts().nlargest(10).plot.pie(figsize=(7,7),label='',autopct='%1.1f%%')
```

```python
dfStocks = client.get_pandas_dataframe(spark,data_set_name='stocks')
dfStocks.set_index('Date')['Close'].plot(figsize=(12,6))
```

```python
dfEmployees = client.get_pandas_dataframe(spark,data_set_name='employees')
dfEmployees['businessUnit'].value_counts().plot(kind='bar',title="Employee Count by Business Unit",figsize=(10,5));
```

```python
import seaborn as sns
sns.pairplot(dfStocks)
```

```python
g = sns.jointplot(x='Volume', y='Close', data=dfStocks, kind="kde", color="b")
g.ax_joint.collections[0].set_alpha(0) 
```
